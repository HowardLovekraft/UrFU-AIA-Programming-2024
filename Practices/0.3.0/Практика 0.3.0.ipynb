# **Задача 4.**




Камень, ножницы, бумага

Создайте игру в Камень, ножницы, бумагу, в которой пользователь может играть против компьютера. Программа должна запросить выбор пользователя, а затем сгенерировать выбор компьютера. Затем программа должна определить победителя на основе правил игры.

**Этап 1:**

Определение победителя

**Этап 2:**
  
Игра до 3 побед


Вввод для "Этап 1":

```
Камень
```

Вывод:

```
Компьютер: Ножницы
Игрок: Камень
Победа: Игрок
```

````python
import random

def rock_paper_scissors(user_fig: str, comp_fig: str) -> str:
    """
        Возрващает имя победителя в партии
        :param user_fig: Ход пользователя
        :param comp_fig: Ход компьютера
        :returns: Имя победителя (Игрок/Компьютер)
    """
    turns = {
        "Камень": "Бумага",
        "Ножницы": "Камень",
        "Бумага": "Ножницы"
    } # value бьёт key

    if user_fig == comp_fig:
        return "Ничья. :)"
    if turns[user_fig] == comp_fig:
        return "Компьютер"
    return "Игрок"


FIGURES = ("Камень", "Ножницы", "Бумага")
user_score, comp_score = 0, 0

try:
    while max(user_score, comp_score) < 3:
        stage = (input(), random.choice(FIGURES)) # unfriendly UI только из-за ТЗ
        print(f"Компьютер: {stage[1]}")
        print(f"Игрок: {stage[0]}")
        winner = rock_paper_scissors(*stage)
        print(f"Победа: {winner}")
        if winner == "Игрок": user_score += 1
        elif winner == "Компьютер": comp_score += 1
        print(f"Счёт (Игрок/Компьютер): {user_score}-{comp_score}\n")

    if comp_score == 3:
        res = "компьютер"
    elif user_score == 3:
        res = "игрок"
    print(f"Победил {res}", end="")
except KeyboardInterrupt:
    print("Ctrl-C by user")
````

# **Задача 7.**



**Цифра на определенном месте:**

Последовательно записан натуральный ряд чисел.

Какая цифра стоит в N позиции


Вввод:

```
Введите номер позиции: 1234567890
```

Вывод:

```
8
```

````python
N = int(input("Введите номер позиции: "))
# >>> 1234567890 -> 8

# Длина ряда N-разрядных чисел - R * 10**R, R - разрядность числа.
# Одноразрядные числа (0-9) - 10 знаков.
# Двуразрядные (10-99) - 200 знаков, etc.

summy_r = 0
r = 0
while summy_r <= N:
    r += 1
    summy_r += r * 10**r

prev_r = r-1
prev_summy = summy_r - r * 10**r

print(summy_r, r)
print(prev_summy, prev_r)
````
