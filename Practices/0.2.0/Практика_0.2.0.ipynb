# Практика 0.2.1

**Шахматы**

Даны стартовые и конечные координаты, а также фигура

Необходимо определить, может ли заданная фигура так ходить?

```python
def filter_crds(crds_list: list[list[int]]) -> list[list[int]]:
    """
    Функция удаляет 'невозможные' позиции из списка позиций.
    :param crds_list: Список координат
    :returns res: 'вычищенный' список возможных позиций
    """
    res = []
    for x, y in crds_list:
        if x < 1 or x > 8 or y < 1 or y > 8:
            continue  # итеративно удалять list'ы с невозможными позициями не удается,
        else: res.append([x, y])  # поэтому такой костыль.
    return res

def turn_checker(possible_coords: list[list[int]], end_crds: list[int]) -> None:
    """
    Функция выводит на экран, может ли ход end_crds быть совершен фигурой, возможными ходами которой являются possible_coords
    :param possible_coords: Возможные ходы фигуры
    :param end_crds: Конечные координаты, введенные пользователем
    """
    return "Такой ход существует" if end_crds in filter_crds(possible_coords) else "Такого хода не существует"


def chess_checker(start_crds: list[int], end_crds: list[int], fig: str) -> str:
    if fig == "пешка":
        possible_crds = [[start_crds[0], start_crds[1]+1]]

    elif fig == "конь":
        possible_crds = []
        for sign_x in -1, 1:
            for sign_y in -1, 1:
                for move_x in 1, 2:
                    for move_y in 1, 2:
                        if move_x == move_y: continue
                        else: possible_crds.append([start_crds[0]+sign_x*move_x, start_crds[1]+sign_y*move_y])

    elif fig == "ладья":
        possible_crds = [[i, start_crds[1]] for i in range(1, 9)] + [[start_crds[0], j] for j in range(1, 9)]

    elif fig == "слон":
        possible_crds = []
        for sign_x in -1, 1:
            for sign_y in -1, 1:
                for step in range(1, 9):
                    possible_crds.append([sign_x*step+start_crds[0], sign_y*step+start_crds[1]])

    elif fig == "королева": # код: ладья+слон
        possible_crds = [[i, start_crds[1]] for i in range(1, 9)] + [[start_crds[0], j] for j in range(1, 9)]
        for sign_x in -1, 1:
            for sign_y in -1, 1:
                for step in range(1, 9):
                    possible_crds.append([sign_x*step+start_crds[0], sign_y*step+start_crds[1]])

    elif fig == "король":
        possible_crds = [[x+start_crds[0], y+start_crds[0]] for x in (-1, 0, 1) for y in (-1, 0, 1)]

    else:
        return "Такой фигуры нет в шахматах :("
    return turn_checker(filter_crds(possible_crds), end_crds)
```

```python
# Почти тесты (без pytest / unittest, но с assert)

assert chess_checker([1, 1], [0, 0], "пешка") == "Такого хода не существует"
assert chess_checker([4, 4], [6, 2], "слон") == "Такой ход существует"
assert chess_checker([4, 4], [6, 2], "шашка") == "Такой фигуры нет в шахматах :("
assert chess_checker([8, 3], [7, 5], "конь") == "Такой ход существует"
```

```python
print("Примечание: координаты A-H в программе заменяются на 1-8. Т.е., например, координату 'A8' в программе нужно представить как '1 8'")
print("Примечание 2: Программа работает только для ходов белых фигур")
start_crds = [int(elem) for elem in input("Введите стартовые координаты через пробел (пример - 1 8 (это 'A8'))): ").split(' ')]
end_crds = [int(elem) for elem in input("Введите конечные координаты через пробел: ").split(' ')]
fig = input("Введите название фигуры на русском (пешка, ладья, конь, слон, король, королева):")

print(chess_checker(start_crds, end_crds, fig))
```

#4 Дополнительная практика

#### Определение корректного расставления скобок
Возьмите код из задания 7 и улучшите ваш калькулятор следующим образом:

Пользователь вводит строку вида:

(5+5)*5 - данная строка содержит в себе математическое выражение, а также скобки

Ваша программа должная понять что за математическое выражение записано, проверить верно ли оно записано (пример неверного заполнения: (5(+)5)*5, а также расчитать его в соотвествии с правилами математики

```python
defi = input("Введите выражение без пробелов (пример - 2+1-(3+1)): ")
def func(defi: str) -> int:
    q = []
    for symb in defi:
        if symb == "(":
            q.append(1)
        elif symb == ")":
            try:
                q.pop()
            except IndexError:
                return -1
    if len(q) == 0:
        return 0
    return -1

res = func(defi)
print("Скобки расставлены правильно" if res == 0 else "Скобки расставлены неправильно")
```
