Создайте объект класса и продемонстрируйте его возможности

````python
class BankAccount:
    """
    Класс, описывающий счёт клиента банка.
    """
    def __init__(self, name: str) -> None:
        """
        Создаёт счёт клиента банка.
        :param name: Имя клиента банка.
        """
        self.name = name
        self.balance = 0.0

    def get_balance(self) -> float:
        """Возвращает информацию о балансе счёта."""
        return self.balance

    def deposit(self, amount: float) -> None:
        """
        Добавляет сумму на счёт.
        :param amount: Сумма для зачисления
        """
        self.balance += amount

    def withdraw(self, amount: float) -> bool:
        """
        Списывает сумму со счёта
        :returns: Итог операции:
            True - списание успешно,
            False - списание не удалось
        """
        current_balance = self.balance - amount
        if current_balance < 0:
            print("На счёте недостаточно средств, транзакция отменена.")
            return False
        self.balance = current_balance
        return True

# реализуй цикл While и интерфейс
````

Задача: Система управления библиотекой

**Цель**
Создайте простую систему управления библиотекой, которая позволит пользователям добавлять книги, брать книги, возвращать книги и просматривать список доступных книг.

**Требования**

1. **Определение класса**:
   – Создайте класс с именем «Book» со следующими атрибутами:
     - `title`
     - `автор`
     - `isbn`
     - `is_borrowed` (по умолчанию `False`)

2. **Класс библиотеки**:
   - Создайте класс с именем Library, который управляет коллекцией книг.
   - Класс должен иметь следующие методы:
     - `__init__(self)`: инициализирует пустой список книг.
     - `add_book(self, book: Book)`: добавляет новую книгу в библиотеку.
     - `borrow_book(self, isbn: str)`: помечает книгу как заимствованную. Если книга не найдена или уже взята, выведите соответствующее сообщение.
     - `return_book(self, isbn: str)`: помечает книгу как возвращенную. Если книга не найдена или не была взята взаймы, выведите соответствующее сообщение.
     - `list_available_books(self)`: печатает список всех доступных книг в библиотеке.
     - `find_book(self, isbn: str)`: возвращает объект книги, если он найден, в противном случае возвращает `None`.

3. **Взаимодействие с пользователем**:
   - Создайте простое текстовое меню, которое позволит пользователям:
     - Добавить книгу
     - Одолжить книгу
     - Вернуть книгу
     - Список доступных книг
     - Выйти из программы

````python
from typing import Optional
from dataclasses import dataclass

# Т.к. класс не содержит каких-либо методов, кроме конструктора
# __init__(), можно определить класс как класс данных.
# Также, т.к. класс в дальнейшем не будет наследоваться/изменяться, можно
# добавить dunder-метод slots для оптимизации траты памяти при
# создании экземпляров класса.
@dataclass(slots=True)
class Book:
    """
    Класс, описывающий книгу в системе управления  библиотекой.
    :param title: Название книги
    :param author: ФИО автора книги
    :param isbn: ISBN книги
    :param is_borrowed: Одолжена ли книга посетителю библиотеки
    """
    title: str
    author: str
    isbn: str
    is_borrowed: bool


class Library():
    """
    Класс, описывающий список книг.
    """
    def __init__(self) -> None:
        """
        Создаёт пустой список книг.
        """
        self.book_list = []

    def add_book(self, book: Book) -> None:
        """
        Добавляет книгу в список книг.
        :param book: Книга, которая добавится в список.
        """
        self.book_list.append(book)

    def find_book(self, isbn: str) -> Optional[Book]:
        """
        Ищет книгу в списке книг.
        :returns: Объект книги, если он найден, иначе - None
        """
        if book in self.book_list:
            return book
        return None

    def borrow_book(self, isbn: str) -> bool:
        """
        Помечает книгу как заимствованную, если она есть в списке книг.
        :param isbn: ISBN книги
        :returns: True, если книгу можно взять, иначе - False
        """
        book = self.find_book(isbn)
        if book is not None and book.is_borrowed = False:
            book.is_borrowed = True
            return True
        return False

    def return_book(self, isbn: str) -  bool:
        book = self.find_book(isbn)
        if book is not None and book.is_borrowed = True:
            book.is_borrowed = False
            return True
        return False

    def list_available_books(self) -> None:
        for book in self.book_list:
            if book.is_borrowed = 0:
                print(book)
````
