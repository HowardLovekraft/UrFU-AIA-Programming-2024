Напишите систему для учёта отпусков с возможностью узнавать, сколько дней отпуска осталось у того или иного сотрудника.
Для этого создайте класс Employee со следующими методами:

- Метод consume_vacation должен отвечать за списание дней отпуска.

Единственный параметр этого метода (кроме self) — количество потраченных отпускных дней (целое число).

При вызове метода consume_vacation соответствующее количество дней должно вычитаться из общего числа доступных отпускных дней сотрудника.

Чтобы определить число доступных отпускных дней конкретного сотрудника, в классе опишите атрибут экземпляра |, который по умолчанию будет равен значению атрибута класса vacation_days, и используйте этот атрибут в работе метода.

- Метод get_vacation_details должен возвращать остаток отпускных дней сотрудника в формате: ```Остаток отпускных дней: <число>.```


Чтобы проверить работу программы:
1. Создайте экземпляр класса Employee.
2. Вызовите метод consume_vacation, указав подходящее значение аргумента, например 7.
3. Вызовите метод get_vacation_details.

````python
class Employee():
    def __init__(self, vacation_days: int = 28):
        self.vacation_days = vacation_days

    def consume_vacation(self, spent_days: int) -> None:
        '''
        Списывает указанное количество отпускных дней из общего числа
        доступных отпускных дней.
        :param spent_days: Количество потраченных отпускных дней
        '''
        setattr(self, 'vacation_days', self.vacation_days - spent_days)

    def get_vacation_details(self) -> None:
        '''
        Печатает оставшееся число отпускных дней сотрудника.
        '''
        print(f'Остаток отпускных дней: {self.vacation_days}')


employee = Employee()
employee.consume_vacation(7)
employee.get_vacation_details()
````

Создайте экземпляр класса Rectangleи продемонстрируйте его функциональность.

````python
class Rectangle:
    def __init__(self, width: int, height: int) -> None:
        """
        Класс, описывающий прямоугольник.
        :param width: Ширина прямоугольника
        :param height: Высота треугольника
        """
        self.width = width
        self.height = height

    def get_area(self) -> int:
        """Возвращает площадь прямоугольника."""
        return self.width * self.height

    def get_perimeter(self) -> int:
        """Возвращает периметр прямоугольника"""
        return 2 * (self.width + self.height)

    def get_params(self) -> tuple[int, int]:
        """Возвращает размеры прямоугольника"""
        return (self.width, self.height)


rectangle = Rectangle(10, 20)
rct_width, rct_height = rectangle.get_params()

print(f'Ширина прямоугольника: {rct_width}, высота: {rct_height}',
      f'Площадь прямоугольника: {rectangle.get_area()}',
      f'Периметр прямоугольника: {rectangle.get_perimeter()}',
      sep='\n')
````

# Задание 2

Задание: Создайте мини версию банковской системы:


Инструкции:

1. Создайте класс BankAccountсо следующими атрибутами:
    - account_holder -  владелец счета
    - balance - баланс счета

2. Реализуйте следующие методы:
    - Метод для инициализации владельца счета: имя владельца счета и установите начальный баланс на 0.
    - deposit(amount): Добавьте указанную сумму к балансу.
    - withdraw(amount): Вычесть указанную сумму из баланса, если средств достаточно; в противном случае вывести предупреждение.
    - get_balance(): Возврат текущего баланса.


Создайте объект класса и продемонстрируйте его возможности

````python
class BankAccount:
    """
    Класс, описывающий счёт клиента банка.
    """
    def __init__(self, account_holder: str) -> None:
        """
        Создаёт счёт клиента банка.
        :param account_holder: Имя владельца счёта.
        """
        self.account_holder = account_holder
        self.balance = 0.0

    def get_balance(self) -> float:
        """Возвращает информацию о балансе счёта."""
        return self.balance

    def deposit(self, amount: float) -> None:
        """
        Добавляет сумму на счёт.
        :param amount: Сумма для зачисления
        """
        self.balance += amount

    def withdraw(self, amount: float) -> bool:
        """
        Списывает сумму со счёта.
        :returns: Итог операции:
            True - списание успешно,
            False - списание не удалось
        """
        current_balance = self.balance - amount
        if current_balance < 0:
            print("На счёте недостаточно средств, транзакция отменена.")
            return False
        self.balance = current_balance
        return True

ACTIONS = {
    '0': 'Проверить баланс счёта',
    '1': 'Пополнить счёт',
    '2': 'Снять деньги со счёта',
    '3': 'Закрыть скрипт'
}
ENUM_ACTIONS = tuple(f'{num} - {action}' for num, action in ACTIONS.items())

print('Вы авторизовались в банковской системе Mistral_Bank.',
      'Чтобы продолжить, введите ваше имя:')
try:
    account_holder = input()
except KeyboardInterupt:
    print('Вы завершили скрипт.')

# Создание экземпляра класса
bank_account = BankAccount(account_holder)
print('Отлично! Доступные операции: ', *ENUM_ACTIONS, sep='\n')

while True:
    action = input('\nВведите номер действия: ')
    if action == '0':
        print(f'Ваш баланс: {bank_account.get_balance()} рублей.')
    elif action == '1':
        try:
            deposit_sum = float(input('Введите сумму для зачисления: '))
        except ValueError:
            print('Вы ввели некорректные символы.')
            continue

        bank_account.deposit(deposit_sum)
        print('Зачисление прошло успешно.',
              f'Ваш баланс: {bank_account.get_balance()} рублей', sep='\n')

    elif action == '2':
        try:
            deposit_sum = float(input('Введите сумму для списания: '))
        except ValueError:
            print('Вы ввели некорректные символы.')
            continue

        if bank_account.withdraw(deposit_sum):
            print('Списание прошло успешно.',
                  f'Ваш баланс: {bank_account.get_balance()} рублей', sep='\n')

    elif action == '3':
        print('Скрипт завершён!')
        break

    else:
        print('Вы ввели неверный номер операции.')
````

````python
import random as rand


class Hero:
    def __init__(self, name, armor, weapon, health):
        '''
        Задаёт характеристики героя.
        :param name: Имя героя.
        :self armor: Количество очков брони героя.
        :self weapon: Диапазон значений урона.
        :self health: Количество очков здоровья героя.
        '''
        self.name = name
        self.armor = armor
        self.weapon = weapon
        self.health = health

    def get_attack_points(self) -> int:
        '''Случайно генерирует количество очков атаки героя.'''
        return rand.choice(self.weapon)

    def get_damage(self, dmg: int) -> int:
        '''
        Наносит урон герою, учитывая поправку на очки брони.
        :param dmg: Урон от противника.
        '''
        armored_dmg = 0 if (dmg - self.armor <= 0) else (dmg - self.armor)
        setattr(self, 'health', (self.health - armored_dmg))
        return self.health


class Dragon:
    def __init__(self, name, health, damage):
        '''
        Задаёт характеристики дракона.
        :param name: Имя дракона.
        :self health: Количество очков здоровья героя.
        :self damage: Диапазон значений урона.
        '''
        self.name = name
        self.health = health
        self.damage = damage

    def get_attack_points(self) -> int:
        '''Случайно генерирует количество очков атаки героя.'''
        return rand.choice(self.damage)

    def get_damage(self, dmg: int) -> int:
        '''
        Наносит урон дракону.
        :param dmg: Урон от противника.
        '''
        setattr(self, 'health', self.health - dmg)
        return self.health


hero, dragon = {'health': 125}, {}

print('Вы авторизовались в Симуляторе битвы рыцаря и дракона.\nЕсли скрипт',
'показывает нумерованный список, вам нужно ввести лишь номер (т.е. без точки).')

hero['name'] = input('Введите имя героя: ')

print('''Вы пробираетесь по лесу в сторону луча света.
Цель всего вашего пути - спасение Принцессы из заточения в башне, которую
охраняет Дракон. На пути вы встречаете милого зайца.
Он навевает вам воспоминания о вашем отрочестве - вы... (выбор доспеха)''')

armor_options = {'были бродягой и воровали.': 3,  # light
                 'были подмастерьем кузнеца и собирали доспехи.': 6,  # medium
                 'выросли в тренировочном лагере воинов.': 15} # heavy

armor_numbers = {
    '1': armor_options['были бродягой и воровали.'],
    '2': armor_options['были подмастерьем кузнеца и собирали доспехи.'],
    '3': armor_options['выросли в тренировочном лагере воинов.']
}

enum_armor_options = (f'{num}.{elem}' for num, elem in zip(armor_numbers,
                                                           armor_options))
print(*enum_armor_options, sep='\n')

while True:
    decision_armor = input('Введите номер: ')
    if decision_armor in armor_numbers:
        hero['armor'] = armor_numbers[decision_armor]
        break
    else:
        print('Введите корректный номер варианта.')

print('''Пробираясь дальше по лесу, вы замечаете деревянный монумент куба d20.
Вы вспоминаете лучшие годы юности - вы...''')

weapon_options = {
    'играли с приятелями в дартс на деньги.': range(7, 12), # bow
    'учились фехтовальному делу.': range(10, 15), # sword
    'работали лесорубом.': range(9, 30) # two-handed axe
}

weapon_numbers = {
    '1': weapon_options['играли с приятелями в дартс на деньги.'],
    '2': weapon_options['учились фехтовальному делу.'],
    '3': weapon_options['работали лесорубом.']
}

enum_weapon_options = (f'{num}.{elem}' for num, elem in zip(weapon_numbers,
                                                            weapon_options))
print(*enum_weapon_options, sep='\n')

while True:
    decision_weapon = input('Введите номер: ')
    if decision_weapon in weapon_numbers:
        hero['weapon'] = weapon_numbers[decision_weapon]
        break
    else:
        print('Введите корректный номер варианта.')

print('''Наконец вы замечаете очертания башни. Ваш заклятый враг - Дракон. Вы пытаетесь
рассмотреть цвет его шкуры и видите, что она ...''')

dragon_options = {
    'зеленая.': (200, range(12, 15), 'Зирус'),  # HP, DMG
    'переливается синим цветом.': (180, range(14, 19), 'Ксилзри'),
    'объята красным пламенем.': (230, range(10, 12), 'Амрием')
}

dragon_numbers = {
    '1': dragon_options['зеленая.'],
    '2': dragon_options['переливается синим цветом.'],
    '3': dragon_options['объята красным пламенем.']
}

enum_dragon_options = (f'{num}.{elem}' for num, elem in zip(dragon_numbers,
                                                            dragon_options))
print(*enum_dragon_options, sep='\n')

while True:
    decision_dragon = input('Введите номер: ')
    if decision_dragon in dragon_numbers:
        dragon_params = dragon_numbers[decision_dragon]
        dragon['health'], dragon['damage'], dragon['name'] = dragon_params
        break
    else:
        print('Введите корректный номер варианта')

print('Да начнется битва!\n')

hero, dragon = Hero(**hero), Dragon(**dragon)

is_hero_wins = True


# Проводит циклы ходов вида Дракон-Герой
while dragon.health > 0:
    dragon_dmg = dragon.get_attack_points()
    hero_health = hero.get_damage(dragon_dmg)
    print(f'{dragon.name} наносит {dragon_dmg} урона. ', end='')

    if hero.health <= 0:
        is_hero_wins = False
        break
    else:
        hero_dmg = hero.get_attack_points()
        dragon_health = dragon.get_damage(dragon_dmg)
        print(f'{hero.name} наносит {hero_dmg} урона.')

if is_hero_wins:
    print('Герой победил!')
else:
    print('Герой побежден :(')
````

# Дополнительное задание


Задача: Система управления библиотекой

**Цель**
Создайте простую систему управления библиотекой, которая позволит пользователям добавлять книги, брать книги, возвращать книги и просматривать список доступных книг.

**Требования**

1. **Определение класса**:
   – Создайте класс с именем «Book» со следующими атрибутами:
     - `title`
     - `автор`
     - `isbn`
     - `is_borrowed` (по умолчанию `False`)

2. **Класс библиотеки**:
   - Создайте класс с именем Library, который управляет коллекцией книг.
   - Класс должен иметь следующие методы:
     - `__init__(self)`: инициализирует пустой список книг.
     - `add_book(self, book: Book)`: добавляет новую книгу в библиотеку.
     - `borrow_book(self, isbn: str)`: помечает книгу как заимствованную. Если книга не найдена или уже взята, выведите соответствующее сообщение.
     - `return_book(self, isbn: str)`: помечает книгу как возвращенную. Если книга не найдена или не была взята взаймы, выведите соответствующее сообщение.
     - `list_available_books(self)`: печатает список всех доступных книг в библиотеке.
     - `find_book(self, isbn: str)`: возвращает объект книги, если он найден, в противном случае возвращает `None`.

3. **Взаимодействие с пользователем**:
   - Создайте простое текстовое меню, которое позволит пользователям:
     - Добавить книгу
     - Одолжить книгу
     - Вернуть книгу
     - Список доступных книг
     - Выйти из программы

````python
from typing import Optional
from dataclasses import dataclass


# Т.к. класс не содержит каких-либо методов, кроме конструктора
# __init__(), можно определить класс как класс данных.
# Также, т.к. класс в дальнейшем не будет наследоваться/изменяться, можно
# добавить dunder-метод slots для оптимизации траты памяти при
# создании экземпляров класса.
@dataclass(slots=True)
class Book:
    """
    Класс, описывающий книгу в системе управления библиотекой.
    :param title: Название книги
    :param author: ФИО автора книги
    :param isbn: ISBN книги
    :param is_borrowed: Одолжена ли книга посетителю библиотеки
    """
    title: str
    author: str
    isbn: str
    is_borrowed: bool = False

    def __repr__(self) -> str:
        return f'Книга: {self.title}\nАвтор: {self.author}' + \
               f'\nISBN: {self.isbn}\nЕсть в библиотеке: ' + \
               f'{"Да" if self.is_borrowed == True else "Нет"}'


class Library():
    """Класс, описывающий список книг."""
    def __init__(self) -> None:
        """Создаёт пустой список книг."""
        self.book_list = []

    def add_book(self, book: Book) -> None:
        """
        Добавляет книгу в список книг.
        :param book: Книга, которая добавится в список.
        """
        self.book_list.append(book)

    def find_book(self, isbn: str) -> Optional[Book]:
        """
        Ищет книгу в списке книг.
        :returns: Объект книги, если он найден, иначе - None.
        """
        for book in self.book_list:
            if book.isbn == isbn:
                return book
        return None

    def borrow_book(self, isbn: str) -> bool:
        """
        Помечает книгу как заимствованную, если она есть в списке книг.
        :param isbn: ISBN книги.
        :returns: True, если книгу можно взять, иначе - False.
        """
        book = self.find_book(isbn)
        if book is not None and book.is_borrowed == False:
            book.is_borrowed = True
            return True
        return False

    def return_book(self, isbn: str) ->  bool:
        """
        Убирает метку заимствования с книги, если она была.
        :param isbn: ISBN книги.
        :returns: True, если книга была взята, иначе - False
        """
        book = self.find_book(isbn)
        if book is not None and book.is_borrowed == True:
            book.is_borrowed = False
            return True
        return False

    def get_list_available_books(self) -> list[Book]:
        """
        Возвращает список доступных книг.
        """
        return [book for book in self.book_list if book.is_borrowed == 0]


ACTIONS = {
    '0': 'Добавить книгу в систему',
    '1': 'Взять книгу',
    '2': 'Вернуть книгу',
    '3': 'Получить список доступных книг',
    '4': 'Завершить работу'
}
ENUM_ACTIONS = tuple(f'{num} - {action}' for num, action in ACTIONS.items())
library = Library()

print('Вы авторизовались в системе DeepMindLibrary. Доступные действия: ',
      *ENUM_ACTIONS, sep='\n')

while True:
    action = input('\nВведите номер операции: ')

    if action == '0':
        title = input('Введите название:\n')
        author = input('Введите ФИО автора:\n')
        isbn = input('Введите ISBN:\n')
        library.add_book(Book(title, author, isbn))
        print('Книга добавлена в систему.')

    elif action == '1':
        isbn = input('Введите ISBN книги:\n')
        borrowed_book = library.borrow_book(isbn)
        if borrowed_book:
            print('Можете забрать книгу у библиотекаря :)')
        else:
            print('К сожалению, книги с таким ISBN в библиотеке сейчас нет.')

    elif action == '2':
        isbn = input('Введите ISBN книги:\n')
        returned_book = library.return_book(isbn)
        if returned_book:
            print('Спасибо за возврат!')
        else:
            print('Вы пытаетесь вернуть книгу, которой нет в системе?',
                  'Или вы хотите дать нам второй экземпляр? :)', sep='\n')

    elif action == '3':
        print('Список доступных книг:',
              *library.get_list_available_books(),
              sep='\n')

    elif action == '4':
        print('Скрипт завершен.')
        break

    else:
        print('Вы ввели неверный номер операции.')
````
